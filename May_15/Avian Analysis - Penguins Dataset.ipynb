**Dataset:** sns.load_dataset('penguins')  
**Goal:** Differentiate penguin species by comparing multiple physical traits at once.

```python
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset("penguins")
```

**Adding Hue, Size, and Style:** Build a sns.scatterplot() comparing bill_length_mm vs. bill_depth_mm. Make it 4-dimensional by assigning hue='species' (different colors for species), style='island' (different marker shapes for locations), and size='body_mass_g' (bubble sizes for weight).

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    style="island",
    size="body_mass_g"
)
plt.title("Penguin Bill Dimensions by Species, Island, and Body Mass", fontsize=14)
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

**Pair Plots:** Run sns.pairplot(penguins, hue='species'). Have the students observe how this one line of code generates a massive grid of scatterplots and KDEs, instantly showing which variables easily separate the Adelie, Chinstrap, and Gentoo penguins.

```python
sns.pairplot(data=penguins, hue="species")
plt.suptitle("Pairwise Relationships and Feature Distributions by Species", y=1.02)
plt.show()
```

**Correlation & Heatmaps:** Have them drop the categorical columns, calculate the correlation matrix (df.corr()), and visualize it using sns.heatmap(annot=True, cmap='coolwarm'). They should discover the strong positive correlation between flipper length and body mass.

```python
numeric = penguins.select_dtypes(include=["float64", "int64"])
correlation_matrix = numeric.corr()
plt.figure(figsize=(8, 6))

sns.heatmap(data=correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Penguin Physical Traits", fontsize=14)
plt.tight_layout()
plt.show()
```
